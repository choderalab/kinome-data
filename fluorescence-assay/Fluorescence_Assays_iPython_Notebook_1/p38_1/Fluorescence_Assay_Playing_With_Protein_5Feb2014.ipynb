## Emission scans wwith 335 nm excitation

```python
fig1 = pyplot.figure(figsize=(12,8))
    
for n in arange(num_wells)+1:
    pyplot.plot(P335_x_num, PROTEIN_1_335[n,1:])
 
pyplot.xlabel('Emission wavelength (nm)')
pyplot.ylabel('Relative Fluorescence')

pyplot.ylim (0,500000)
pyplot.xlim (280,700)

pyplot.title('OVERVIEW:Em scan with Clear Corning plates, ex=335nm, 280-850nm, step=2, BW=5, Bottom Read, Gain=100')
pyplot.legend(legend1)

pyplot.savefig('P335_1_Overview')
pyplot.show()
```

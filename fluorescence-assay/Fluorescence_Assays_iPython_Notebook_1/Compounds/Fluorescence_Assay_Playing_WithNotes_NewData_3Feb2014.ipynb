## Absorbance Scans

```python
fig1 = pyplot.figure(figsize=(12,8))
    
for n in arange(num_wells)+1:
    pyplot.plot(ABx_num, ABS_BLACK[n,1:])
 
pyplot.xlabel('Absorbance wavelength (nm)')
pyplot.ylabel('Absorbance Magnitude')

pyplot.ylim (0,4.5)

pyplot.title('Absorbance scan with Black Corning plates, 230-900nm, step=5, 20 flashes')
pyplot.legend(legend1)

pyplot.savefig('NewData_AB1')
pyplot.show()
```

```python
fig2 = pyplot.figure(figsize=(12,8))
legend2=('blank','dI water','DMSO','QZA','ERL','Aniline Blue')
    
for n in arange(1,7):
    pyplot.plot(ABx_num, ABS_BLACK[n,1:])

pyplot.xlabel('Absorbance wavelength (nm)')
pyplot.ylabel('Absorbance Magnitude')

pyplot.ylim (0,4.5)

pyplot.axvline(280,linewidth=2.5,linestyle='dashed',color='k')
pyplot.axvline(340,linewidth=2.5,linestyle='dashed',color='k')
pyplot.axvline(380,linewidth=2.5,linestyle='dashed',color='k')

pyplot.title('Absorbance scan with Black Corning plates, 230-900nm, step=5, 20 flashes')
pyplot.legend(legend2)

pyplot.savefig('NewData_AB2')
pyplot.show()
```

#### ** Absorbance scan highlights: 

##### - Triplicate replicate show little variation, though some is apparent, likely due to pipetting error. This error is most obvious in absorbance scans using the clear plates. Worth checking what this variation is like when using Tecan Liquid Handling.

```python
HERE = Image(filename='NewData_AC1.PNG')
HERE
```

##### - Blank / dI water show absorbance at < 280 nm in Black plates but not Clear plates.

##### - DMSO shows absorbance at < 300 nm in Black plates but not Clear plates.

```python
fig_ab_highlights_1 = pyplot.figure(figsize=(12,8))

for n in arange(1,4):
    pyplot.plot(ABx_num, ABS_BLACK[n,1:])

for n in arange(1,4):
    pyplot.plot(ACx_num, ABS_CLEAR[n,1:],linestyle='dashed')
    
pyplot.xlabel('Absorbance wavelength (nm)')
pyplot.ylabel('Absorbance Magnitude')
legend3=('blank-black','dI water-black','DMSO-black','blank-clear','dI water-clear','DMSO-clear')

pyplot.ylim (0,4.5)
pyplot.xlim (200,400)

pyplot.axvline(280,linewidth=2.5,linestyle='dashed',color='k')
pyplot.axvline(340,linewidth=2.5,linestyle='dashed',color='k')
pyplot.axvline(380,linewidth=2.5,linestyle='dashed',color='k')

pyplot.title('CONTROLS: Absorbance scan with Clear and Black Plates, 230-900nm, step=5, 20 flashes')
pyplot.legend(legend3)

pyplot.savefig('NewData_AB_HIGHLIGHTS_1')
pyplot.show()
```

##### - Absorbance spectra markedly different in magnitude, though not necessarily in shape for ERL/QZA/Aniline Blue in two different kinds of plates.

```python
fig_ab_highlights_2 = pyplot.figure(figsize=(12,8))

for n in arange(4,7):
    pyplot.plot(ABx_num, ABS_BLACK[n,1:])

for n in arange(4,7):
    pyplot.plot(ACx_num, ABS_CLEAR[n,1:],linestyle='dashed')
    
pyplot.xlabel('Absorbance wavelength (nm)')
pyplot.ylabel('Absorbance Magnitude')
legend3=('QZA-black','ERL-black','Aniline Blue-black','QZA-clear','ERL-clear','Aniline Blue-clear')

pyplot.ylim (0,4.5)
pyplot.xlim (250,600)

pyplot.axvline(280,linewidth=2.5,linestyle='dashed',color='k')
pyplot.axvline(340,linewidth=2.5,linestyle='dashed',color='k')
pyplot.axvline(380,linewidth=2.5,linestyle='dashed',color='k')

pyplot.title('COMPOUNDS: Absorbance scan with Clear and Black Plates, 230-900nm, step=5, 20 flashes')
pyplot.legend(legend3)

pyplot.savefig('NewData_AB_HIGHLIGHTS_2')
pyplot.show()
```

# Let's consider some data for a fluorescence assay prepared with a Tecan HP D300 digital dispenser.

A summary of the experimental procedure:

>>
Two protein assay plates were prepared with p38 and Src plus  Bosutinib, using rows A-E and columns 1-12, varying protein concentration from row to row (high to low where E is just Buffer)  and Bosutinib concentration from column to column (high to low where 12 has no Bosutinib). 
Bosutinib concentrations started at 20 uM and went down by ½ log as calcuated by the D300 dispenser. Src concentrations were ~1, 0.5, 0.25, and 0.125 uM.  p38 concentrations were 10, 5, 2.5, 1.25 uM.

```python
import re
import numpy as np
from matplotlib.pyplot import *
%pylab inline
```

```python
# Load data
SRC_280 = np.genfromtxt('12022014-019-src-280scan-reboot-bottom.txt',dtype='str')
SRC_280_x = SRC_280[0,:]
SRC_280_x_num = re.findall(r'\d+', str(SRC_280_x )[1:-1])
```

```python
# Plot the emission spectrum for 280 nm excitation
legend1=(SRC_280[1:,0])
num_wells=len(legend1)

rcParams['figure.figsize'] = [8, 6]
for n in np.arange(num_wells)+1:
    plot(SRC_280_x_num, SRC_280[n,1:])

axis([300, 800, 0, 5000]);
xlabel('emission wavelength (nm)');
ylabel('fluorescence units');
```

```python
# Find desired emission wavelength to fit.
wavelength_to_fit = '480'
index = SRC_280_x_num.index(wavelength_to_fit)

# Specify row to process
row_index = 0 # row 0, 1, 2, or 3
Pstated_rows = numpy.array([1, 0.5, 0.25, 0.125]) * 1e-6 # protein concentrations, M

# Extract fluorescence intensities at desired wavelength.
F_i = np.array([ float(SRC_280[i,index]) for i in range(row_index*12+1,row_index*12+13) ])
Fligand_i = np.array([ float(SRC_280[i,index]) for i in range(49,61) ])

# Stated concentrations of protein and ligand.
Pstated = Pstated_rows[row_index] # protein concentration, M
Lstated = 20.0e-6 / np.array([10**(float(i)/2.0) for i in range(12)])
```

```python
# Plot fluorescence intensity vs ligand concentration.
fig2 = pyplot.figure(figsize=(10,5))
rcParams['figure.figsize'] = [10, 5];
clf();
semilogx(Lstated, Fligand_i, 'ro');
xlabel('$[L]_T$ (M)');
ylabel('rel. fluorescence');
```

```python
# Plot fluorescence intensity vs ligand concentration on a log plot to see if we are running into inner filter effects.
fig2 = pyplot.figure(figsize=(10,5))
rcParams['figure.figsize'] = [10, 5];
clf();
loglog(Lstated, Fligand_i, 'ro');
xlabel('$[L]_T$ (M)');
ylabel('rel. fluorescence');
```

```python
import pymc

# Two-component binding.
def two_component_binding(DeltaG, P, L):
    Kd = np.exp(-DeltaG)
    PL = 0.5 * ((P + L + Kd) - np.sqrt((P + L + Kd)**2 - 4*P*L));  # complex concentration (M)                                                                                                                                                                                                         
    P = P - PL; # free protein concentration in sample cell after n injections (M)                                                                                                                                                                                                                          
    L = L - PL; # free ligand concentration in sample cell after n injections (M)                                                                                                                                                                                                                           
    return [P, L, PL]

# Create a pymc model
def make_model(Lstated, dLstated, Fobs_i, Fligand_i):
    N = len(Lstated)
    
    # Priors on true concentrations of protein and ligand.
    #Ptrue = pymc.Lognormal('Ptrue', mu=np.log(Pstated**2 / np.sqrt(dPstated**2 + Pstated**2)), tau=np.sqrt(np.log(1.0 + dPstated**2/Pstated**2))**(-2)) # protein concentration (M)
    #Ltrue = pymc.Lognormal('Ltrue', mu=np.log(Lstated**2 / np.sqrt(dLstated**2 + Lstated**2)), tau=np.sqrt(np.log(1.0 + dLstated**2/Lstated**2))**(-2)) # ligand concentration (M)
    Ltrue = pymc.Normal('Ltrue', mu=Lstated, tau=dLstated**(-2)) # ligand concentration (M)

    # Priors on fluorescence intensities of complexes (later divided by a factor of Pstated for scale).
    #F_background = pymc.Gamma('F_background', alpha=0.1, beta=0.1, value=Fobs_i.min()) # background fluorescence
    Fmax = max(Fobs_i.max(), Fligand_i.max())
    F_background = pymc.Uniform('F_background', lower=0.0, upper=Fmax) # background fluorescence
    F_L = pymc.Uniform('F_L', lower=0.0, upper=Fmax/Lstated.max()) # ligand fluorescence
    
    # extinction coefficient
    epsilon = pymc.Uniform('epsilon', lower=0.0, upper=100000, value=70000) # extinction coefficient for ligand
    
    # Unknown experimental measurement error.
    log_sigma = pymc.Uniform('log_sigma', lower=-3, upper=+3, value=0.0) 
    @pymc.deterministic
    def precision(log_sigma=log_sigma): # measurement precision
        return 1.0 / np.exp(log_sigma)**2

    well_volume = 75e-6 # L
    well_area = 0.1586 # cm^2 # half-area
    path_length = well_volume * 1000 / well_area # cm
 
    # Fluorescence model, ligand only.
    @pymc.deterministic
    def Fligand(F_background=F_background, F_L=F_L, Ltrue=Ltrue, epsilon=epsilon):
        Fmodel_i = np.zeros([N])
        for i in range(N):
            IF = 1 - numpy.exp(-epsilon*path_length*Ltrue[i])
            #Fmodel_i[i] = (F_L*Ltrue[i] + F_background)*IF # weird different attempt at inner filter effect (WRONG) (DON'T USE)
            #Fmodel_i[i] = F_L*Ltrue[i] + F_background # WITHOUT inner filter effect
            Fmodel_i[i] = F_L*IF + F_background # WITH inner filter effect
        return Fmodel_i
    
    # Experimental error on fluorescence observations.
    Fligand_model = pymc.Normal('Fblank_i', mu=Fligand, tau=precision, size=[N], observed=True, value=Fligand_i) # ligand only data
    
    
    # Construct dictionary of model variables.
    pymc_model = { 'Ltrue' : Ltrue, 
                   'log_sigma' : log_sigma, 
                   'precision' : precision, 
                   'F_L' : F_L, 
                   'F_background' : F_background,
                   'Fligand_i' : Fligand,
                   'Fligand_model' : Fligand_model, 
                   'epsilon' : epsilon # extinction coefficient for ligand
                   }
    return pymc_model

# Uncertainties in protein and ligand concentrations.
dLstated = 0.08 * Lstated # ligand concentraiton uncertainty (due to gravimetric preparation and HP D300 dispensing)

# Build model.
pymc_model = pymc.Model(make_model(Lstated, dLstated, F_i, Fligand_i))

# Sample with MCMC
mcmc = pymc.MCMC(pymc_model, db='ram', name='Sampler', verbose=True)
mcmc.sample(iter=40000, burn=20000, thin=50, progress_bar=False)
```

```python
# Plot trace of true ligand concentration.
rcParams['figure.figsize'] = [15, 3]
plot(mcmc.Ltrue.trace(), 'o');
xlabel('MCMC sample');
ylabel('$[L]$ (M)');
```

```python
# Plot trace of intrinsic fluorescence parameters.
rcParams['figure.figsize'] = [15, 3]
semilogy(mcmc.F_L.trace(), 'o', mcmc.F_background.trace(), 'o');
legend(['ligand fluorescence', 'background fluorescence']);
xlabel('MCMC sample');
ylabel('relative fluorescence intensity');
```

```python
# Plot model fit.
pyplot.rcParams['figure.figsize'] = [15, 3]
Fmodels = mcmc.Fligand_i.trace()
clf();
hold(True)
for Fmodel in Fmodels:
    semilogx(Lstated, Fmodel, '-')
semilogx(Lstated, Fligand_i, 'ro')
hold(False)
xlabel('$[L]_s$ (M)');
ylabel('fluorescence units');
```

```python
# Plot trace of measurement error.
rcParams['figure.figsize'] = [15, 3]
plot(np.exp(mcmc.log_sigma.trace()), 'o');
xlabel('MCMC sample');
ylabel('fluorescence measurement error');
```

```python
# Plot extinction coefficient
rcParams['figure.figsize'] = [15, 3]
plot(mcmc.epsilon.trace(), 'o');
xlabel('MCMC sample');
ylabel('ligand extinction coefficient');
```
